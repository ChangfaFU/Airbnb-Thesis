```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('http://data.insideairbnb.com/switzerland/vd/vaud/2019-02-18/visualisations/listings.csv',header=0,sep=',',index_col=0)
df.head(5)
```

```python
print(df.shape)
```

```python
df2=df['neighbourhood'].value_counts()
df2.to_csv('df3')
df4=pd.read_csv('df3',header=None)
df4.columns=['neighbourhood','Count']
display(df4)
```

```python
import folium
latitude = 46.62
longitude = 6.55
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=9)
```

```python
world_geo = r'http://data.insideairbnb.com/switzerland/vd/vaud/2019-02-18/visualisations/neighbourhoods.geojson'
threshold_scale = np.linspace(df4['Count'].min(),
                              df4['Count'].max(),
                              6, dtype=int)
threshold_scale = threshold_scale.tolist() 
threshold_scale[-1] = threshold_scale[-1] + 1 
sanfran_map.choropleth(
    geo_data=world_geo,
    data=df4,
    columns=['neighbourhood','Count'],
    key_on='feature.properties.neighbourhood',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Airbnb in Vaud'
)

sanfran_map
```
